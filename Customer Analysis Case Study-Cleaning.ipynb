## Cleaning and wrangling

```python
#find the unique values in the columns we want to change
print("State -> ", df["State"].unique())
print("Gender -> ", df["Gender"].unique())
print("Education ->", df["Education"].unique())
print("Customer Lifetime Value ->", df["Customer Lifetime Value"].unique())
print("Income ->", df["Income"].unique())
print("Monthly Premium Auto ->", df["Monthly Premium Auto"].unique())
print("Number of Open Complaints ->", df["Number of Open Complaints"].unique())
print("Policy Type ->", df["Policy Type"].unique())
print("Vehicle Class ->", df["Vehicle Class"].unique())
print("Total Claim Amount ->", df["Total Claim Amount"].unique())
```

```python
#rename same values under one name
df['State'] = df['State'].replace(['AZ', 'Cali', 'WA'],['Arizona', 'California', 'Washington'])
df
df['State'].value_counts()
```

```python
#rename same values under one name
df['Gender'] = df['Gender'].replace(['Femal', 'female', 'Male'],['F', 'F', 'M'])
df
df['Gender'].value_counts()
```

```python
#rename same values under one name
df['Education'] = df['Education'].replace(['Bachelor'],['Bachelors'])
df
df['Education'].value_counts()
```

```python
df.info()
```

```python
#remove percentage mark 
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].apply(lambda x: float(x.strip('%')) / 100 if isinstance(x, str) else x)
df['Customer Lifetime Value'] 
```

```python
#change the Customer Lifetime Value data typet to float
df['Customer Lifetime Value'] =  pd.to_numeric(df['Customer Lifetime Value'], errors='coerce')
df.info()
df['Customer Lifetime Value'] = df['Customer Lifetime Value'].apply(lambda x: float(round(x, 0)) if isinstance(x, (int, float)) else x)
df['Customer Lifetime Value']
```

```python
#df['Number of Open Complaints'] = df['Number of Open Complaints'].apply(lambda x: x.split('/') if isinstance(x, str) else x)
#df['Number of Open Complaints'] 
```

```python
#split the Number of Complaints to reach the actual numbers
df['Number of Open Complaints'] = df['Number of Open Complaints'].apply(lambda x: float(x.split('/')[1]) if isinstance(x, str) else x)
df['Number of Open Complaints']
```

## Activity 2 (Tuesday)
-Replacing null values – Replace missing values with means of the column (for numerical columns). Pay attention that the Income feature for instance has 0s which is equivalent to null values. (We assume here that there is no such income with 0 as it refers to missing values) Hint: numpy.nan is considered of float64 data type.

-Bucketing the data - Write a function to replace column "State" to different zones. California as West Region, Oregon as North West, and Washington as East, and Arizona and Nevada as Central

-(Optional) Standardizing the data – Use string functions to standardize the text data (lower case)

```python
#Replacing the null values with mean
df["Customer Lifetime Value"] =df[["Customer Lifetime Value"]].fillna(value=df["Customer Lifetime Value"].mean())
df
```

```python
#This is the original mean of the income column
income_mean = df["Income"].mean()
income_mean
```

```python
#Filled the nan columns with the original mean
df["Income"] = df["Income"].fillna(income_mean)
df
```

```python
#replaced the 0.0 with the original mean
df["Income"] = df["Income"].replace(0,income_mean)
df
```

```python
#Replaced null values with mean
df["Monthly Premium Auto"] =df[["Monthly Premium Auto"]].fillna(value=df["Monthly Premium Auto"].mean())
df
```

```python
#replaced null values with Median
nc_median = df['Number of Open Complaints'].median()
nc_median
df["Number of Open Complaints"] =df[["Number of Open Complaints"]].fillna(value=df["Number of Open Complaints"].median())
df
```
